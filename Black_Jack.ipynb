```python
#BLACK JACK
# Blackjack Game Rules
# Objective:
# The goal of Blackjack is to have a hand value as close to 21 as possible, without exceeding it. The player plays against the dealer (computer), and the hand closest to 21 wins.
# Card Values:
# Ace (A): Can be worth 1 or 11, depending on what is more favorable for the player.
# 2 to 10: Face value (i.e., 2 is worth 2 points, 3 is worth 3 points, etc.).
# Face Cards (King, Queen, Jack): Each worth 10 points.
# Gameplay:
# Betting Phase:

# The player starts by placing a bet using virtual chips.
# Dealing Cards:

# Both the player and the dealer receive two cards. The player's cards are visible, but one of the dealer's cards is face down (hidden).
# Player's Turn:

# The player is asked to choose between:
# Hit: Draw an additional card.
# Stand: Keep current cards and end the turn.
# Double Down: Double the bet, draw one more card, and then stand automatically.
# Fold: Stop the round and forfeit the bet.
# The player continues to take actions until they either decide to stand, go over 21 (bust), or reach 21 (Blackjack).
# Dealer's Turn:

# After the player stands, the dealer reveals their hidden card.
# The dealer must hit (draw cards) until they reach a minimum of 17 points.
# If the dealer exceeds 21, the player wins.
# Winning Conditions:

# If the player's hand value is greater than the dealer's hand without going over 21, the player wins.
# If both hands are equal, it's a tie (push).
# If the player goes over 21, they lose, even if the dealer also busts.
# If the dealer goes over 21, the player wins.
# Blackjack:

# If the player gets an Ace and a 10-point card (10, Jack, Queen, King) as their initial two cards, they have a "Blackjack" and win immediately unless the dealer also has a Blackjack, in which case itâ€™s a tie.
# End of Game:
# The game continues as long as the player has chips to bet. If the player runs out of chips, the game ends.

import random as rd
def deal_card():
    cards=[11,2,3,4,5,6,7,8,9,10,10,10,10]
    card=rd.choice(cards)
    return card
# you can add moeny to m=number of chips using maps
def black_jack(chips):
    money=0
    count=chips
    bet=int(input("Please enter the amount of money you want to bet.\n"))
    q=bet
    user_cards=[]
    computer_cards=[]
    for i in range(2):
        user_cards.append(deal_card())
        computer_cards.append(deal_card())
    a=sum(user_cards)
    print(f'Computer cards are, {computer_cards[0]}, *.\n')
    call=True
    while(call):
        print(f'Your cards are {user_cards}. Total: {a}\n')
        i=input("Click 'Yes' to add another card, 'Double' to make double,or 'stand' to stop dealing.\n").lower()
        if i=="yes" and a<21:
            user_cards.append(deal_card())
            print(f'Your cards are {user_cards}.Total: {a}\n')
            a=sum(user_cards)
            count=chips
            if a>21:
                money=chips-bet
                print(f'Your total crossed 21, so you lost. You have {money} chips in your pocket.\n')
                return ask_user(money)
        elif i=="double" and a<21:
            call=False
            user_cards.append(deal_card())
            a=sum(user_cards)
            print(f'Your cards are {user_cards}.Total: {a}\n')
            if a>21:
                money=count-2*q
                print(f'Your total crossed 21, so you lost. You have {money} chips in your pocket.')
                return ask_user(money)
            count=chips-2*q
            q=2*bet
        elif i=="stand":
            call=False
            count=chips
            a=sum(user_cards)
            print("Its dealer turn to take the cards.\n")
    
    if (a <= 21):
        b=sum(computer_cards)
        if b>=17:
            print(computer_cards)
        while(b<17):
            computer_cards.append(deal_card())
            b=sum(computer_cards)
            print(f'Dealer cards are,{computer_cards}. Total: {b}\n')
        print(b)
        print(a)
        if b>21:
            print("Dealer got busted!!")
            money=count+(2*q)
            print(f'You won, now you have {money} in your pocket.\n')
        elif (b==a):
            print(f'Your total is {a}, and computer total is {b}.Its a Tie!!\n')
            money=count
            print(f'You have {money} in your pocket.\n')
        elif a==21:
            print(f'You hit BLACK JACK.You WON!!\n')
            money=count+(2*q)
            print(f'You have {money} in your pocket.\n')
        elif (a>b):
            print(f'Your total is {a}, and computer total is {b}.YOU WON !!')
            money=count+(2*q)
            print(f'You have {money} in your pocket.\n')
        elif b==21:
            print(f'Dealer hit BLACK JACK.You LOST!!\n')
            money=chips-q
            print(f'You have {money} in your pocket.\n')
        else:
            print(f'Your total is {a}, and computer total is {b}.YOU LOST!!\n')
            money=chips-q
            print(f'You have {money} in your pocket.')
        return ask_user(money)
        

def ask_user(money):
    k=input("Do you want to play one more round? 'y' for 'yes' and 'n' for 'No'.\n").lower()
    if k=="y":
        return black_jack(money)
    else:
        print("Thank you for playing.\n")
        return money

money=int(input("Welcome to BLACK JACK!! Please enter the amount of chips you want to play with, 1$ = 1 chip."))
chips=black_jack(money)
print(f'Your final winning is {chips}.\n')

```
